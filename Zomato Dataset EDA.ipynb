### Zomato Dataset EDA 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## importing the data
df=pd.read_csv('zomato.csv',encoding="latin-1")
df.head()
```

```python
#data specifications
df.shape
```

```python
#Basic info
df.info()
```

```python
# columns
df.columns
```

```python
df.describe()
```

```python
#missing values columns
df.isnull().sum()
```

```python
#missing value 
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
# seaborn heatmap
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
#reading excel
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head(2)
```

```python
#To check data types
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_values=final_df.Country.value_counts().values
```

```python
#Pie chart for top 3 countries uses Zomato
plt.pie(country_values[:3],labels=country_names[:3])
```

### Observations:Top 3 countries are as follows
* India
* United States
* United Kingdom

```python
#Pie chart for top 3 countries uses Zomato with percentage with Percentage
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.1f%%')
```

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
# Renamaing the "0" column to "Rating count"
final_df.groupby(['Aggregate rating', 'Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating count"})
```

```python
rating=final_df.groupby(['Aggregate rating', 'Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating count"})
```

```python
rating
```

## Observation
1. When Rating is between 4.5 to 4.9.....Excellent
2. When Rating is between 4.0 to 3.4.....Very Good
3. When Rating is between 3.5 to 3.9.....Good

```python
#matplotlib.rcParams Plot 
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color',data=rating,palette=['white','red','orange','yellow','green','green'])
```

Observation:
    1.Not Rated is high in number
    2.Max rating are in range of 2.5 to 3.4

```python
sns.countplot(x="Rating color",data=rating,palette=['white','red','orange','yellow','green','green'])
```
